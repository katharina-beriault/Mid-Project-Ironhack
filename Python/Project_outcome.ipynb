#### Categorical data

**ChiSquare test:** <br>
Test to check if categorical variables are independent from each other!
- used to determine if there is a significant relationship between two nominal (categorical) variables
- works with nominal categorical variables well and not well ordinal categorical variables as the test is based on a contingency table and the order in which frequencies are put in the table, does not change the result of the chi-square test
- in the contingency table, the frequency of each category for one nominal variable is compared across the frequencies of categories of the second nominal variable

*Steps for ChiSquare test*
1. Creating a contingency table
2. Checking the shape of the contingency table
3. Calculating the degrees of freedom
4. Calculating ChiSquare-value
5. Comparing ChiSquare-value with critical value from ChiSquare-test-table (for needed p-value) 
6. Result:
    - ChiSquare-value is less than critical value (e.g. for p-value 0.05) on ChiSquare-test-table = the null hypothesis (H0) can be rejected 
    - if H0 can be rejected = there is a correlation between both variables

```python
# contingency table
data_crosstab = pd.crosstab(data['offer_accepted'], data['reward'], margins = False)
data_crosstab
```

### Dealing with data imbalance: Class Weigths

Logistic Regression has an in-built function which can be used to change the class weights, but KNN-Algorithm doesn't which is why this strategy will only be used and tested with the Logistic Regression model.

```python
total = sum(data['offer_accepted'].value_counts())

weights = {'Yes': total/(2*data['offer_accepted'].value_counts()['Yes']),
           'No': total/(2*data['offer_accepted'].value_counts()['No'])}
```
